```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

from sklearn import datasets
iris = datasets.load_iris()
iris_X = iris.data[iris.target!=2, 0:2]
iris_y = iris.target[iris.target!=2]

def visualize(model, name, X, y, area=None, with_scatter=True, with_grid=True):
    """model による分類結果と決定境界を可視化する．
    描画領域 area は (x_min, ymin, x_max, y_max) のタプルとして指定する．
    area が未指定のときは自動設定する．
    """
    X = np.array(X)  # 入力 X や y がリストでもよいように，内部で変換する．
    y = np.array(y)
    
    model.fit(X, y)
    accuracy = model.score(X, y)
    
    cmap = ListedColormap(('red', 'blue'))
    
    if area:
        x_min, y_min, x_max, y_max = area
    #areaが未指定なので自動設定する
    else:     
        x_min, x_max = X[:,0].min() - 1, X[:,0].max() + 1
        y_min, y_max = X[:,1].min() - 1, X[:,1].max() + 1

    xx = np.linspace(x_min, x_max, 300)
    yy = np.linspace(y_min, y_max, 300)
    x_mesh, y_mesh = np.meshgrid(xx, yy)
    z = model.predict(np.array([x_mesh.ravel(), y_mesh.ravel()]).T).reshape(x_mesh.shape)
    
    fig, ax = plt.subplots()
    ax.contourf(xx, yy, z, alpha=0.1, cmap=cmap, vmin=0, vmax=1)

    if with_scatter:
        ax.scatter(X[:,0], X[:,1], c=y, cmap=cmap, vmin=0, vmax=1) 
    ax.set_title(f"{name}, accuracy = {accuracy}")
    ax.set_xlim(x_min, x_max)
    ax.set_ylim(y_min, y_max)
    ax.set_aspect('equal')    # 縦と横の目盛りの幅を揃える．
    if with_grid:
        ax.grid(color='gray', linestyle='dashed', linewidth=0.1)
        ax.set_axisbelow(True)
    fig.show()
```

# 新しいセクション

```python
from sklearn.neighbors import KNeighborsClassifier
model_k1 = KNeighborsClassifier(n_neighbors = 1)     # Nearest Neighbor

```

image.png

```python
area = (0, 0, 10, 10)

#練習問題　決定境界が上のようになるデータを見つける
X = np.array([[2, 2], [8, 8]])
y = np.array([0, 1])

visualize(model_k1, "KNeighborsClassifier (k=1)", X, y, area)
```
